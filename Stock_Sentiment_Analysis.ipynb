```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import plotly.graph_objects as go
```

```python
#Example
tcs = yf.Ticker("TCS.NS")
```

```python
hist = tcs.history(period="max")
```

```python
hist["Open"].plot(figsize=(15, 5), title="TCS Stock Price")
plt.show()
```

```python
stonks = [
    "RELIANCE.NS",
    "WIPRO.NS",
    "HDFCBANK.NS",
    "TCS.NS",
    "INFY.NS",
    "O",
    "MELI",
    "ISRG",
    "DIS",
    "BRK-B",
    "AMZN",
    "ZM",
    "PFE",
    "CLX",
    "DPZ",
    "RTX",
]
```

```python
hists = {}
for s in stonks:
    tkr = yf.Ticker(s)
    history = tkr.history(period="4y")
    hists[s] = history
```

# Plot the Stock Price

```python
for stock in stonks:
    temp_df = hists[stock].copy()

    fig = go.Figure(
        data=[
            go.Candlestick(
                x=temp_df.index,
                open=temp_df["Open"],
                high=temp_df["High"],
                low=temp_df["Low"],
                close=temp_df["Close"],
            )
        ]
    )

    fig.update_layout(
        margin=dict(l=20, r=20, t=60, b=20),
        height=1000,
        paper_bgcolor="LightSteelBlue",
        title=stock,
    )

    fig.show()
```
